```python
from sklearn.cluster import SpectralClustering
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
RANDOM_STATE = 42
```

```python
df = pd.read_csv('./data/t48k.csv')
```

```python
df.head()
```

```python
X = df.to_numpy()
```

```python
fig = plt.figure(2)
plt.scatter(X[:,0], X[:,1], color='blue', marker='.', label='targets')
plt.show()
```

```python
num_class =6
```

```python
clusteringHier = AgglomerativeClustering(n_clusters= num_class).fit_predict(X)
```

```python
data_listSpi = []
clusters = clusteringSpi
for i in range(num_class):
  tmp = np.argwhere(clusters == i)
  id = tmp.reshape((tmp.shape[0],))
  xTmp = X[id]
  data_listSpi.append(xTmp)
```

```python
data_listHier = []
clusters = clusteringHier
for i in range(num_class):
  tmp = np.argwhere(clusters == i)
  id = tmp.reshape((tmp.shape[0],))
  xTmp = X[id]
  data_listHier.append(xTmp)
```

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo','purple']
fig, ax = plt.subplots(1, 1)
data_listX = data_listSpi
for i in range(num_class):
  ax.scatter(data_listX[i][:,0], data_listX[i][:,1], color=colors[i], marker='.', label= 'class '+str(i) )
  ax.set_xlabel('$x$')
  ax.set_ylabel('$y$')
  ax.legend()
plt.show()
```

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo','purple']
fig, ax = plt.subplots(1, 1)
data_listX = data_listHier
for i in range(num_class):
  ax.scatter(data_listX[i][:,0], data_listX[i][:,1], color=colors[i], marker='.', label= 'class '+str(i) )
  ax.set_xlabel('$x$')
  ax.set_ylabel('$y$')
  #ax.legend()
plt.show()
```
